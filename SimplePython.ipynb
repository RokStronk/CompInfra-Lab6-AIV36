```python
name = input("Enter your name:")
while True:
    try:
        age = int(input("Enter your age:"))
    except:
        print("Thats not a number...")
    else:
        break;

print("Hello {0}! Your age is {1}!".format(name, age))
birth_year = 2026 - age
print("You were born around {0}.".format(birth_year))
```
